```python
import pandas as pd

data_unclean_filepath = "data-unclean.csv"
data_clean_filepath = "data-clean.csv"
```

```python
data = pd.read_csv(data_unclean_filepath)

pd.set_option("display.max_columns", None)
data.head()
```

# Fixing DtypeWarning: Columns (15) have mixed types

```python
data = pd.read_csv(data_unclean_filepath, dtype={"address": str})

print(f"COUNT OF TYPES BEFORE:\n{data["address"].apply(type).value_counts()}\n------------------------")

problematic_rows = data[~data["address"].apply(lambda address: isinstance(address, str))]
data["address"] = data["address"].fillna("Unknown") # CLEANING

print(f"COUNT OF TYPES AFTER:\n{data["address"].apply(type).value_counts()}")
```

# Dropping columns

```python
# dropping
data = data.drop(["title", "body", "amenities", "has_photo", "price_display", "source"], axis=1)

date = pd.to_datetime(data["time"], unit="s").dt.strftime("%m/%d/%Y")
time = pd.to_datetime(data["time"], unit="s").dt.strftime("%H:%M %p")
data = data.drop("time", axis=1)
data["date"] = date
data["time"] = time

data.head()
```

# Renaming columns

```python
# renaming
data = data.rename(columns={
    "pets_allowed": "pets allowed",
    "price_type": "price type",
    "square_feet": "square feet",
    "cityname": "city"
})

data.head()
```

# Changing NaN and Dtype values

```python
data.info()
```

```python
data = data.fillna({
    "bathrooms": "Unknown",
    "bedrooms": "Unknown",
    "pets allowed": "Unknown",
    "price": "Unknown",
    "city": "Unknown",
    "state": "Unknown",
    "latitude": "Unknown",
    "longitude": "Unknown",
})
```

```python
 data[data["pets allowed"] == "Cats,Dogs,None"]
```

```python
# --> data["INSERT_COLUMN_HERE"].apply(type).value_counts()
# --> data[~data["INSERT_COLUMN_HERE"].apply(lambda entry: isinstance(entry, INSERT_DESIRED_TYPE_HERE))].head()

# dropping rows
data = data.drop(index=[83369, 6864])

# strings
string_columns = data.select_dtypes(include="object").columns[data.select_dtypes(include="object").apply(lambda column: column.map(type).eq(str).all())]
data[string_columns] = data[string_columns].astype("string")

# floats
data["price"] = data["price"].astype("float")

data["longitude"] = data["longitude"].apply(lambda longitude: longitude.replace(".", "", 1)) # Removing unnecessary . from entries
data["longitude"] = data["longitude"].apply(lambda longitude: float(longitude) if longitude != "Unknown" else longitude) # Cast entries to float
```
